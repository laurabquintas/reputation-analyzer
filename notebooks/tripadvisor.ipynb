# Tripadvisor Debug Notebook

Step-by-step checks for `src/sites/tripadvisor.py`.

Requires `TRIPADVISOR_API_KEY` environment variable.

```python
from pathlib import Path
import sys
import os

ROOT = Path.cwd()
if not (ROOT / 'src').exists() and (ROOT.parent / 'src').exists():
    ROOT = ROOT.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

print('Project root:', ROOT)
```

```python
from src.sites import tripadvisor as site
```

```python
api_key = os.getenv('TRIPADVISOR_API_KEY')
if not api_key:
    raise RuntimeError('Set TRIPADVISOR_API_KEY before running this notebook')
```

```python
site.LOCATION_IDS
```

```python
hotel, location_id = next(iter(site.LOCATION_IDS.items()))
print('Testing:', hotel, location_id)
score, num_reviews = site.ta_get_rating(location_id, api_key=api_key)
print('Score:', score, '| Reviews:', num_reviews)
```

```python
results = {}
for hotel, location_id in site.LOCATION_IDS.items():
    try:
        score, num_reviews = site.ta_get_rating(location_id, api_key=api_key)
        results[hotel] = {'score': score, 'num_reviews': num_reviews}
    except Exception as exc:
        results[hotel] = {'error': str(exc)}
results
```

```python
from datetime import datetime
import subprocess

date_col = datetime.now().strftime('%Y-%m-%d')
cmd = [sys.executable, str(ROOT / 'src' / 'sites' / 'tripadvisor.py'), '--date', date_col]
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=False)
```
